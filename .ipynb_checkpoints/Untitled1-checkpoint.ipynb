```python
import numpy as np
import random
import matplotlib.pyplot as plt

# Создаем пустой массив 5x5
sprite = np.zeros((5,5))

# Задаем случайные значения для каждого пикселя
for i in range(5):
    for j in range(3):
        sprite[i,j] = random.randint(0,1)
        sprite[i,4-j] = sprite[i,j]

# Отображаем спрайт
plt.imshow(sprite, cmap='gray')
plt.show()
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt

# Создаем пустую матрицу 25x25, где каждый элемент - спрайт 5x5 пикселей
x = 10
y = 5
sprites = np.zeros((x * 5 + x, y * 5 + y,5,5))

# Задаем случайные значения для каждого спрайта
for i in range(x):
    for j in range(y):
        for k in range(5):
            for l in range(3):
                sprites[i*5+i,j*5+j,k,l] = random.randint(0,1)
                sprites[i*5+i,j*5+j,k,4-l] = sprites[i*5+i,j*5+j,k,l]

# Объединяем все спрайты в одну карту спрайтов
sprite_map = np.block([[sprites[i,j] for j in range(25)] for i in range(25)])

# Отображаем карту спрайтов
plt.imshow(sprite_map, cmap='gray')
plt.show()
```
